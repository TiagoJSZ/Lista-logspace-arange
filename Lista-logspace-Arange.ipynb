```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as pl
```

```python
np.logspace(2, 8, num=20,base=4)
```

```python
x = np.logspace(0.1, 20)
y = x*x
```

```python
pl.plot(x,y,"c")
pl.plot(x,y,"bo")
pl.xlabel("DIAS")
pl.ylabel("INFECTADOS")
pl.title("MODELO MATEMÁTICO PARA PREVISÃO DA CURVA EPIDEMIOLÓGICA DE COVID-19 NO BRASIL")
pl.grid()
plt.figure(figsize =(16,20))
```
